### 2) Differentiation & Integration for Circular Motion
Consider a point mass undergoing uniform circular motion of radius $R$ in the $xy$-plane, with constant angular velocity $\omega$. The parametric equations of motion are:

$$
\begin{aligned}
x(t) &= R \cos(\omega t),\\
y(t) &= R \sin(\omega t).
\end{aligned}
$$

1. **Velocity and Acceleration**:  
   - Symbolically differentiate $x(t)$ and $y(t)$ with respect to $t$ to find the velocity components $v_x(t)$ and $v_y(t)$.  
   - Differentiate again to find the acceleration components $a_x(t)$ and $a_y(t)$.  


2. **Integration Check**:
